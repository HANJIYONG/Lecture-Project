```
vocab={}
sentences=[]
s=0
for i in b:
    result=[]
    for j in i:
        result.append(i)
        if(j in vocab.keys()):
            pass
        else:
            s+=1
            vocab[j]=s
    sentences.append(result)
```

```
encoding=[]
for i in b:
    encod=[]
    for j in range(len(i)):
        encod.append(vocab[i[j]])
    encoding.append(encod)
print(encoding[:100])
```

```
lenlist=[]
for i in encoding:
    lenlist.append(len(i))
print(max(lenlist))
max_len=max(lenlist)
```

```
v_size=len(vocab)+1
print(v_size)
```

```
from keras.models import Sequential
from keras.preprocessing.sequence import pad_sequences

# max_len=1588

data = pad_sequences(encoding, maxlen=max_len, padding='post', value=0)
```

```
print(data)
```

```
y=np.array(y)
```

```
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(data,y, test_size=0.2,random_state=42)
```

```
print(type(x_train[0]))
print(type(y_train[0]))
print(type(x_test[0]))
print(type(y_test[0]))
```

```
from keras.layers import Input, Embedding, Conv1D, Concatenate, GlobalMaxPooling1D, Dense,LSTM
from keras.models import Model
```

```
input_data = Input(shape=(max_len,))
print(input_data)
emb = Embedding(v_size, 128)(input_data)
print(emb)

cnn_3 = Conv1D(256, 3, strides=1, padding='same')(emb)
print(cnn_3)
cnn_4 = Conv1D(256, 4, strides=1, padding='same')(emb)
print(cnn_4)

# lstm1=LSTM(256,return_sequences=True)(emb)

cnn_5 = Conv1D(256, 5, strides=1, padding='same')(emb)
print(cnn_5)

concat = Concatenate(axis=-1)([cnn_3, cnn_4, cnn_5])
print(concat)




maxpooling = GlobalMaxPooling1D()(concat)
print(maxpooling)



final = Dense(1)(maxpooling)
print(final)
```

```
model = Model(inputs=input_data, outputs=final)
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae','mse'])
model.summary()
```

```
#예측 

example_batch = x_train[:10]
example_result = model.predict(example_batch)
example_result
```

```
from keras.callbacks import EarlyStopping
```

```
early_stopping = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
```

```
hist = model.fit(x_train, y_train,
                 epochs=30,
                 batch_size=256,
                 validation_split=0.2,
                 callbacks=[early_stopping])
```

```
print(hist.history['loss'])
print(hist.history['mean_squared_error'])
```

```
def predict_score(summary):
    for i in range(len(summary)):
      summary = re.sub("(<br\s*/><br\s*/>)|(\-)|(\/)", " ", summary)
      summary = re.sub("[.;:!\'?,\"()\[\]]", ' ', summary)

    token = kma.morphs(summary)

    enc=[]
    for i in range(len(token)):
      try:
        enc.append(vocab[token[i]])
      except KeyError:
        enc.append(vocab[token[i-1]])


    max_len=1588
    enc_data = pad_sequences([enc], maxlen=max_len, padding='post', value=0)


    
    predict=model.predict(enc_data)
    print("[{}]\n---> 예상 평점 {:.2f}점입니다\n".format(summary, predict[0][0]))
    # return predict[0][0]
```

```
import matplotlib.pyplot as plt

def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mean_absolute_error'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mean_absolute_error'],
           label = 'Val Error')
  plt.ylim([0,18])
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mean_squared_error'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mean_squared_error'],
           label = 'Val Error')
  plt.ylim([0,18])
  plt.legend()
  plt.show()

# plot_history(history)
```
